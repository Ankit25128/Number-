Extract 3rd column

Extract 2×2 submatrix (center)

```python
A = np.arange(1, 26).reshape(5,5)
print(A)

# Extract 2nd row
print("Extract 2nd row")

secnd_row = A[1]
print(secnd_row)

# Extract 3rd column
print("Extract 3rd colmn")
thrd_colmn=A[:,2]
print(thrd_colmn)

# Extract 2x2 center matrix
print("2x2 matrix")
centr = A[2:4,2:4]
print(centr)


# Reverse last row
print("reversing last Row")
rev_last_row = A[-1][::-1]
print(rev_last_row)
```
